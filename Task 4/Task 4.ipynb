## Exploratory Data Analysis

```python
plt.figure(figsize = (14, 8))
sns.barplot(x = 'Ship Mode', y = 'Profit', data = data, hue = 'Region')
plt.show()
```

From the graph above, we can say that Ship Mode 'Same Day' in Region 'South' there is negative profit. It means the company is getting loss in South region and company has to focus more on south region. 

```python
plt.figure(figsize = (14, 8))
sns.barplot(x = 'Ship Mode', y = 'Profit', data = data, hue = 'Segment')
plt.show()
```

Almost every time, 'Home Office' Segment is giving more profit when compared to other Segments and 'Consumer' Segment giving less profit.

```python
plt.figure(figsize = (14, 8))
sns.barplot(x = 'Region', y = 'Profit', data = data, hue = 'Segment')
plt.show()
```

In the 'East' Region and 'Home Office' Segment is giving high profits. 'Central' Region and 'Consumer' Segment is giving very low profits.

```python
plt.figure(figsize = (14, 8))
sns.barplot(x = 'Category', y = 'Profit', data = data)
plt.show()
```

```python
plt.figure(figsize = (14, 8))
sns.barplot(x = 'Sub-Category', y = 'Profit', data = data)
plt.show()
```

```python
plt.figure(figsize = (14, 8))
sns.barplot(x = 'Category', y = 'Profit', data = data, hue = 'Sub-Category')
plt.show()
```

From the above graph, Copiers are giving very high profits.

```python
state_profit = data.groupby('State')['Profit'].mean()

plt.figure(figsize = (20, 10))
plt.pie(x = state_profit, labels = state_profit.index, shadow = True)
plt.show()
```

From the above pie chart, Vermot City is giving high profits

```python
plt.figure(figsize = (20, 10))
sns.scatterplot(x = 'Sales', y = 'Profit', data = data, color = 'g')
plt.show()
```

For profit > 0, sales are almost linearly increasing. 

```python
plt.figure(figsize = (20, 10))
sns.scatterplot(x = 'Discount', y = 'Profit', data = data, color = 'y')
plt.show()
```

For less discounts, there is more Profit.

```python
plt.figure(figsize = (20, 10))
sns.scatterplot(x = 'Discount', y = 'Sales', data = data, color = 'b')
plt.show()
```

```python
plt.figure(figsize = (20, 10))
sns.scatterplot(x = 'Quantity', y = 'Profit', data = data, color = 'b', hue = 'Category')
plt.show()
```

```python
plt.figure(figsize = (20, 10))
sns.scatterplot(x = 'Quantity', y = 'Profit', data = data, color = 'b', hue = 'Sub-Category')
plt.show()
```

```python
plt.figure(figsize = (20, 8))
sns.distplot(data['Sales'], kde = False)
plt.show()
```

```python
plt.figure(figsize = (20, 8))
sns.distplot(data['Profit'], kde = False)
plt.show()
```

From the data analysis. I found that,                                                                                           
1. Region 'South' with Ship Mode 'Same Day' is giving negative profits, company has to focus on the Region 'South'.             
2. Segment 'Home Office' is getting more profits.                                                                               
3. We are getting more profits from 'Vermont' City.                                                                             
4. 'Technology' is giving more profit especially 'Copiers' are giving very high profit.                                         
5. Less Discounts are giving more Profit.
